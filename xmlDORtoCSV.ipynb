```python
import xml.etree.ElementTree as ET 
import os
import glob

filepaths = glob.glob(os.path.join("./data", "*.xml"))

print(len(filepaths))
```

```python
import string
printable = set(string.printable)

def printRecur(root):
    """Recursively prints the tree."""
    if root.text != None and '\n' not in root.text:
        dataDict[root.tag] = root.text
    for elem in root.getchildren():
        printRecur(elem)

data = []
for files in filepaths:
    try:
        with open(files, 'rb') as f:
            fileIn = f.read()
        fileIn = fileIn.decode(errors='ignore').encode('ascii',errors='ignore').decode()
        root=ET.fromstring(fileIn)
        dataDict = {}
        printRecur(root)
        data.append(dataDict)
    except:
        print("File Couldn't Be Opened:", files)

        
```

```python
import pandas as pd 

df = pd.DataFrame(data)

df.head()
```
