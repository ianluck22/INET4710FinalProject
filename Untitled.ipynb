```python
import pandas as pd 

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras import applications
```

```python
df = pd.read_csv("imageFilenames.csv")

df.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
df.drop(["a"], axis=1, inplace=True)

df = df[df["imgFilepath"] != "no photo"]

df.drop_duplicates("mediumImageLink", keep = "first", inplace=True)

df = df[df["totPurchaseAmt"] > 5000]     # Minimum House Price 5000
df = df[df["totPurchaseAmt"] < 5000000]  # Maximum House Price 5 Million
```

```python
scaled_df = df[["totPurchaseAmt", "imgFilepath"]]

# scaled_df["totPurchaseAmt"] = scaled_df["totPurchaseAmt"] / max(scaled_df["totPurchaseAmt"])
```

```python
IMG_SIZE = 128

def resizeAndLoadData(data):
    imageData = []
    for i, row in data.iterrows():
        try:
            img = Image.open(row["imgFilepath"])
            img = img.resize((IMG_SIZE, IMG_SIZE))
            imageData.append(np.array(img))
        except:
            imageData.append(None)
    data["imgData"] = imageData

resizeAndLoadData(scaled_df)

scaled_df = scaled_df[scaled_df["imgData"].notna()]
```

```python
X = np.array([i for i in np.array(scaled_df["imgData"])]).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
y = scaled_df["totPurchaseAmt"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
plt.imshow(X[10])
```

```python
vggModel = Sequential()

vggModel.add(applications.VGG16(weights="imagenet", input_shape=(IMG_SIZE, IMG_SIZE, 3), include_top=False))

for layer in vggModel.layers:
    layer.trainable = False

vggModel.add(Flatten())
vggModel.add(Dropout(.2))
vggModel.add(Dense(128, activation="relu"))
vggModel.add(Dense(1, activation = 'linear'))

vggModel.summary()
```

```python
vggModel.compile(loss='mean_absolute_percentage_error', optimizer='adam', metrics = ['mae'])
```

```python
vggModel.fit(X_train, y_train, batch_size=50, epochs=100, verbose=1,  validation_split=0.1)
```
