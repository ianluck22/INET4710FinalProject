```python
## Getting rid of any columns were more than 40% null values. 

NA_val = data.isna().sum()
def na_filter(na, threshold = .4): 
#only select variables that passees the threshold
  col_pass = []
  for i in na.keys():
      if na[i]/data.shape[0]<threshold:
          col_pass.append(i)
  col_passdata = data[na_filter(NA_val)]
  return col_passdata
data.columns
```
